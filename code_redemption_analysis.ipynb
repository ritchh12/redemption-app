# User Behavior & Fraud Analysis in Code Redemption System

## Objective
Analyze user behavior and detect fraud patterns in a code redemption system to improve efficiency and security.

```python
import pandas as pd

df = pd.read_csv('attempts_data.csv')
df.head()
```

## Data Overview

```python
df.info()
df.describe()
```

## Data Cleaning

```python
df['attempt_time'] = pd.to_datetime(df['attempt_time'])
df['hour'] = df['attempt_time'].dt.hour
```

## Exploratory Data Analysis

```python
# Success vs Failure Rate
df['is_success'].value_counts(normalize=True) * 100
```

```python
# Failure Reasons
df[df['is_success']==False]['reason'].value_counts()
```

```python
# Peak Usage Time
df['hour'].value_counts().sort_index()
```

```python
# Top Active Users
df['user_id'].value_counts().head(10)
```

```python
# Fraud Detection (Failed Attempts)
df[df['is_success']==False].groupby('user_id').size().sort_values(ascending=False)
```

```python
# Success Rate per User
df.groupby('user_id')['is_success'].mean().sort_values()
```

```python
# Attempts Over Time
df.set_index('attempt_time').resample('H').size()
```

## Visualization

```python
import matplotlib.pyplot as plt

# Success vs Failure
df['is_success'].value_counts().plot(kind='bar', title='Success vs Failure')
plt.show()

# Failure Reasons
df[df['is_success']==False]['reason'].value_counts().plot(kind='bar', title='Failure Reasons')
plt.show()

# Activity by Hour
df['hour'].value_counts().sort_index().plot(kind='bar', title='Usage by Hour')
plt.show()
```

## Key Insights
- Majority of attempts are successful
- Invalid codes are the main failure reason
- Peak activity occurs in evening
- Few users contribute to most failures
- Fraud users show repeated failed attempts

## Conclusion
The system performs efficiently overall, but some users exhibit suspicious behavior. Improving validation and adding stricter controls can enhance reliability.
